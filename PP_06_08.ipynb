```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Upload the Titanic dataset

data = pd.read_csv("titanic (2).csv")


data
```

```python
# 1. Identify the deck with the most passengers
data['Deck'] = data['Cabin'].str[0]  # Extract the deck from the Cabin column
deck_counts = data['Deck'].value_counts()
deck_with_most_passengers = deck_counts.idxmax()
print(f"Deck with the most passengers: {deck_with_most_passengers}")
```

```python
# 2. Calculate the average age of passengers in each class
average_age_by_class = data.groupby('Pclass')['Age'].mean()
print("\nAverage age of passengers by class:")
print(average_age_by_class)
```

```python
3
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='Fare', hue='Survived', fill=True, common_norm=False, alpha=0.5)
plt.title('Density Plot of Fare by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
passengers_southampton = data[data['Embarked'] == 'S'].shape[0]
print(f"\nNumber of passengers who embarked from Southampton: {passengers_southampton}")
```

```python
c_deck_data = data[data['Deck'] == 'C']
c_deck_survival_rate = c_deck_data['Survived'].mean()
print(f"\nSurvival rate for passengers in the 'C' deck: {c_deck_survival_rate:.2f}")
```
